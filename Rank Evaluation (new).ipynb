```python
import Orange
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import friedmanchisquare
```

# Ablation Study
## Spring Simulation
### Recall

```python
## Friedman Chisquare Test
file_path = "./results/rankAblation.xlsx"
df = pd.read_excel(file_path, index_col=None, 
                  usecols="A:D", sheet_name="SpringRecall")
df = df.dropna()
df
```

```python
nri_sym = df["NRI-Sym"].values
wavenetnri_uns = df["WavenetNRI-Uns"].values
nri = df["NRI"].values
wavenetnri = df["WavenetNRI"].values
friedmanchisquare(nri_sym, wavenetnri_uns, nri, wavenetnri)
```

```python
## Recall
ax = plt.figure()

names = ["NRI:1.878", "NRI-Sym:3.617", "WavenetNRI-Uns:1.233", 
        "WavenetNRI:3.272"]

avranks = [1.878, 3.617, 1.233, 3.272]

cd = Orange.evaluation.compute_CD(avranks, 90 , alpha="0.05", test="nemenyi")
print("cd=",cd)

Orange.evaluation.graph_ranks(avranks, names, cd=cd, width=6, textspace=1.)


plt.show()
```

# Experiments
## Spring Simulation
### Recall

```python
## Friedman Chisquare Test
file_path = "./results/rankExperiments.xlsx"
df = pd.read_excel(file_path, index_col=None, 
                  usecols="A:G", sheet_name="SpringRecall")
df = df.dropna()
df
```

```python
nri = df["NRI"].values
nriun = df["NRI(un)"].values
wavenetnri = df["WavenetNRI"].values
wavenetnriun = df["WavenetNRI(un)"].values
GDGAN = df["GD-GAN"].values
Yamaguchi = df["Yamaguchi"].values
solera = df["Solera"].values
friedmanchisquare(nri, nriun, wavenetnri, wavenetnriun,
                 GDGAN, Yamaguchi, solera)
```

```python
#recall
names = ["NRI:1.089", "NRI(un):5.444", 
         "WavenetNRI:2.089",
         "WavenetNRI(un):5.644)", 
         "GD-GAN:4.578", "Yamaguchi:4.528", 
         "Solera:4.628"]

avranks = [1.089, 5.444 , 2.089, 5.644, 4.578, 4.528, 4.628]

cd = Orange.evaluation.compute_CD(avranks, 90 , alpha="0.05")
print("cd=",cd)

Orange.evaluation.graph_ranks(avranks, names, cd=cd, width=5, textspace=1.)
plt.show()
```
